```python
# 원하는 지역을 입력받아 시각화
import csv
import matplotlib.pyplot as plt

area = input("원하는 지역 입력")

f = open('c:/sh/study/study everyday/data files/age.csv',encoding='cp949')
data = csv.reader(f)
next(data)
result = []
for row in data:
    if area in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',',"")))
plt.style.use('ggplot')
plt.rc('font',family = 'Malgun Gothic')
plt.title(area+' 인구 구조 선 그래프')
plt.plot(result)
plt.show()
```

```python
# 남/녀 인구 데이터 저장
import csv
male = []
female = []
f = open('c:/sh/study/study everyday/data files/gender.csv')
data = csv.reader(f)
next(data)
for row in data:
    if '충현동' in row[0]:
        for i in range(0,101):
            male.append(int(row[i+3]))
            female.append(int(row[-(i+1)]))
female.reverse()
print(male)
print('\n')
print(female)
```

```python
# 다른 방법으로 데이터 저장
import csv
male = []
female = []
f = open('c:/sh/study/study everyday/data files/gender.csv')
data = csv.reader(f)
next(data)
for row in data:
    if '충현동' in row[0]:
        for i in row[3:104]:
            male.append(int(i))
        for i in row[106:]:
            female.append(int(i))
print(male)
print('\n')
print(female)
```

```python
# 항아리 모양 그래프 그리기
import matplotlib.pyplot as plt
import csv
male = []
female = []
f = open('c:/sh/study/study everyday/data files/gender.csv')
data = csv.reader(f)
next(data)
for row in data:
    if '충현동' in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:]:
            female.append(int(i))
plt.rc('font',family = 'Malgun Gothic')
plt.title('충현동 남/여 인구구조 막대 그래프')
plt.barh(range(101),male, color = 'pink')
plt.barh(range(101),female, color = 'skyblue')
plt.show()
```

```python
# 원하는 지역의 항아리 그래프 출력
import matplotlib.pyplot as plt
import csv
male = []
female = []
area = input()
f = open('c:/sh/study/study everyday/data files/gender.csv')
data = csv.reader(f)
next(data)
for row in data:
    if area in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:]:
            female.append(int(i))
        break
plt.rc('font',family = 'Malgun Gothic')
plt.title(area+' 남/여 인구구조 막대 그래프')
plt.barh(range(101),male, color = 'pink',label = '남자')
plt.barh(range(101),female, color = 'skyblue', label = '여자')
plt.legend()
plt.show()
```

```python
#제주도 남여 성별 비율
import matplotlib.pyplot as plt
import csv
area = input()
size = []
f = open('c:/sh/study/study everyday/data files/gender.csv')
data = csv.reader(f)
next(data)
for row in data:
    if area in row[0]:
        male = 0
        female = 0
        for i in range(0,101):
            m = row[i+3]
            m = int(m.replace(",",""))
            male += m
            f = row[i+106]
            f = int(f.replace(",",""))
            female += f
        break
size.append(male)
size.append(female)
plt.rc('font',family = 'Malgun Gothic')
plt.title(area+' 성별 비율 그래프')
plt.axis('equal')
plt.pie(size,labels = ['남','여'], autopct = '%.1f%%', startangle=90)
plt.show()
```

```python
# 인구 구조를 꺾은선 그래프로 표현하기
import matplotlib.pyplot as plt
import csv
area = input()
size = []
male = []
female = []
f = open('c:/sh/study/study everyday/data files/gender.csv',encoding='cp949')
data = csv.reader(f)
next(data)
for row in data:
    if area in row[0]:
        for i in range(0,101):
            m = row[i+3]
            m = int(m.replace(",",""))
            male.append(m)
            f = row[i+106]
            f = int(f.replace(",",""))
            female.append(f)
        break
size.append(male)
size.append(female)
plt.rc('font',family = 'Malgun Gothic')
plt.title(area+' 연령별 남/여 인구비율 그래프')
plt.plot(male, label = '남성')
plt.plot(female, label = '여성')
plt.legend()
plt.show()
```

```python
# 성별 차이 막대그래프로 표현
import matplotlib.pyplot as plt
import csv
area = input()
result = []
f = open('c:/sh/study/study everyday/data files/gender.csv')
data = csv.reader(f)
next(data)
for row in data:
    if area in row[0]:
        for i in range(0,101):
            m = row[i+3]
            m = int(m.replace(",",""))
            f = row[i+106]
            f = int(f.replace(",",""))
            result.append(m-f)
        break
size.append(male)
size.append(female)
plt.rc('font',family = 'Malgun Gothic')
plt.title(area+' 남/여 연령별 인구수 차이 막대 그래프')
plt.bar(range(101),result,color = 'skyblue')
plt.show()
```
